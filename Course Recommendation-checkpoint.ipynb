```python
import pandas as pd
import neattext.functions as nfx
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity,linear_kernel
```

```python
df = pd.read_csv('Coursera.csv')
df.head()
```

```python
df['Course Name']
```

```python
df['clean_Course Name'] = df['Course Name'].apply(nfx.remove_stopwords)
```

```python
df['clean_Course Name'] = df['clean_Course Name'].apply(nfx.remove_special_characters)
```

```python
df[['Course Name','clean_Course Name']]
```

```python
count_vect = CountVectorizer()
cv_mat = count_vect.fit_transform(df['clean_Course Name'])
```

```python
cv_mat.todense()
```

```python
df_cv_words = pd.DataFrame(cv_mat.todense(),columns=count_vect.get_feature_names())
```

```python
df_cv_words.head()
```

```python
cosine_sin_mat = cosine_similarity(cv_mat)
```

```python
cosine_sin_mat
```

```python
import seaborn as sns
sns.heatmap(cosine_sin_mat[0:10],annot=True)
```

```python
course_indices = pd.Series(df.index,index=df['Course Name']).drop_duplicates()
```

```python
course_indices
```

```python
course_indices['Retrieve Data using Single-Table SQL Queries']
```

```python
idx = course_indices['Retrieve Data using Single-Table SQL Queries']
```

```python
idx
```

```python
scores = list(enumerate(cosine_sin_mat[idx]))
```

```python
scores
```

```python
sorted_scores = sorted(scores,key=lambda x:x[1],reverse=True)
```

```python
sorted_scores[1:]
```

```python
selected_course_indices = [i[0] for i in sorted_scores[1:]]
```

```python
selected_course_indices
```

```python
selected_course_scores = [i[1] for i in sorted_scores[1:]]
```

```python
recommended_result = df['Course Name'].iloc[selected_course_indices]
```

```python
rec_df = pd.DataFrame(recommended_result)
```

```python
rec_df
```

```python
rec_df['similarity_scores'] = selected_course_scores
```

```python
rec_df
```

```python
def recommend_course(title,num_of_rec=10):
    idx = course_indices[title]
    scores = list(enumerate(cosine_sin_mat[idx]))
    sorted_scores = sorted(scores,key=lambda x:x[1],reverse=True)
    selected_course_indices = [i[0] for i in sorted_scores[1:]]
    selected_course_scores = [i[1] for i in sorted_scores[1:]]
    result = df['Course Name'].iloc[selected_course_indices]
    rec_df = pd.DataFrame(result)
    rec_df['similarity_scores'] = selected_course_scores
    return rec_df.head(num_of_rec)
```

```python
recommend_course('SQL for Data Science')
```
